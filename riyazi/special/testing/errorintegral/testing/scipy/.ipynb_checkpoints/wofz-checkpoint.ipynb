### Faddeeva function

```python
from scipy import special
import matplotlib.pyplot as plt
```

```python
import numpy as np 
x = np.linspace(-3, 3)
z = special.wofz(x)
```

```python
plt.plot(x, z.real, label='wofz(x).real')
plt.plot(x, z.imag, label='wofz(x).imag')
plt.xlabel('$x$')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```

```python
from cmath import exp
from scipy import special
def wofz(z):
    return exp(-z**2) * special.erfc(-1j*z)
```

```python
special.wofz(3)
```
