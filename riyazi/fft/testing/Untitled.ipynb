```python
import numpy as np
import random

exp = np.exp
pi = np.pi

def mydft(X):
    n = len(X)
    out = []
    for k in range(n):
        temp = 0
        for i in range(n):
            temp += X[i] * exp(-2j*pi*k*i/n)
        out.append(temp)
    return np.array(out)

def myidft(X):
    n = len(X)
    out = []
    for k in range(n):
        temp = 0
        for i in range(n):
            temp += X[i] * exp(2j*pi*k*i/n)
        out.append(temp)
    return (1/n) * np.array(out)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.fftpack import fft

# create samples
data = np.zeros(1024)
for k in range(0, 1024):
    data[k] = np.sin(k/20 * np.pi*2)

# plot original samples
#plt.plot(data)
#plt.show()

# apply FFT on samples
waves = fft(data)

# plot FFT result
#plt.plot(np.imag(waves),'r')
#plt.plot(np.real(waves),'b')
#plt.ylabel('value')
#plt.xlabel('period')
#plt.show()

# 
res = np.zeros(1024)
for k in range(0, 1024):
    val = 0.0
    for n in range(0, len(waves)-1):
        # https://dsp.stackexchange.com/a/510/25943
        val += waves[n]*np.exp(-1.j * 2*np.pi * n * k / len(waves)) / len(waves)
    res[k] = val.real


#np implementation
res2 = np.fft.ifft(waves)

plt.plot(data, 'b') # original
plt.plot(res,'g') # my implementation
plt.plot(res2,'y') # np implementation
plt.show()
```
