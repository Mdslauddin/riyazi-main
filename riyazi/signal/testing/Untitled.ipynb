## To plot sina and cos wave

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# to define indepedent variable 
t =np .arange(0,1,0.001)
xt = np.sin(2*np.pi*3*t)
```

```python
# to plot the singal 
plt.plot(t,xt)
plt.title("Sine Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

### To plot sampled sin and cos signal 

```python
# define independent and dependent variable 

n = np.arange(0,1,0.001)
xn = np.sin(2*np.pi*3*n)
```

```python
# to plost the singal 

plt.stem(n,xn)
plt.title("Samples sin Wave")
plt.xlabel("------>n")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np

def sine_wave(frequency, amplitude, phase, sampling_rate, duration):
    # Generate time values
    time = np.linspace(0, duration, int(sampling_rate * duration))
    # Generate sine wave values
    sine_wave = amplitude * np.sin(2 * np.pi * frequency * time + phase)
    # to plot the singal 
    plt.plot(sine_wave)
    plt.title("Sine Wave")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.show()
```

```python
import numpy as np

def cosine_wave(frequency, amplitude, phase, sampling_rate, duration):
    # Generate time values
    time = np.linspace(0, duration, int(sampling_rate * duration))
    # Generate cosine wave values
    cosine_wave = amplitude * np.cos(2 * np.pi * frequency * time + phase)
    # to plot the singal 
    plt.plot(cosine_wave)
    plt.title("Sine Wave")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.show()
```
